Housing price Predictor Model (Linear Regression)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn import datasets
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()

housing_df = pd.DataFrame(data=housing.data, columns=housing.feature_names)
housing_df['HousePrice'] = housing.target
housing_df.head()
```

Now let's plot scatter plot to better understand the relationship between HousePrice and medicalInsurance

```python
medInc = housing_df['MedInc']
houseprice = housing_df['HousePrice']

plt.figure(figsize=(8,6))
plt.scatter(medInc, houseprice, alpha=0.5, color='green')
plt.title('Scatter Plot of MedInc vs HousePrice')
plt.xlabel('MedInc')
plt.ylabel('HousePrice')
plt.grid(True)
plt.show()
```

Let's calculate the Varaiance Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

housing_df_vif = housing_df.drop('HousePrice', axis=1)
housing_df_vif = housing_df_vif.apply(pd.to_numeric, errors='coerce')

vif_data = pd.DataFrame()
vif_data["feature"] = housing_df_vif.columns
vif_data['VIF'] = [variance_inflation_factor(housing_df_vif.values, i) for i in range(len(housing_df_vif.columns))]

print(vif_data)
```

Obersavtion: Latitude and Longitude have a very high value compare to others. Thus, removing or adjusting these values is necessary for right intepretation and understanding.

```python
housing_df = housing_df.drop(['Latitude', 'Longitude'], axis=1)
```

```python
#Let convert some of the data to numeric

cols_to_convert = ['MedInc', 'HouseAge', 'AveBedrms', 'Population', 'AveOccup']
housing_df[cols_to_convert] = housing_df[cols_to_convert].apply(pd.to_numeric, errors='coerce')
housing_df['HousePrice'] = pd.to_numeric(housing_df['HousePrice'], errors='coerce')

x= sm.add_constant(housing_df.drop('HousePrice', axis=1))
y= housing_df['HousePrice']

model = sm.OLS(y, x).fit()
```

Now that we have numeric data, Let's perform residual analysis to validate Linear regression assumptions and visualise the distribution residuals

```python
residuals = model.resid

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
sns.histplot(residuals, kde=True)
plt.title('Residuals Distribution')

plt.subplot(1, 2, 2)
plt.scatter(model.predict(x), residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals vs Predicted Values')

plt.ylabel('Residuals')
plt.show()

print("Model coefficient:\n", model.params)
```

Let's compute and interprete the R-squared and Mean Square Error and to also visualise

```python
from sklearn.metrics import mean_squared_error, r2_score

predictions = model.predict(x)

print("R-squared: ", r2_score(y, predictions))
print("MSE:", mean_squared_error(y, predictions))
```

```python
import matplotlib.pyplot as plt

def plot_boxplots(df):

  fig, axes = plt.subplots(1, len(df.columns), figsize=(20, 5))

  for i, column_name in enumerate(df.columns):
    df.boxplot(column=column_name, ax=axes[i])
    axes[i].set_title(column_name)

  plt.show()

plot_boxplots(housing_df)
```

Observation: We noticed outlier in the MedInc, Population and Houseprice. To ensure efficiency we need to remove the outliers.

```python
def remove_ouliers(df):
  df_no_ouliers = df.copy()

  for column_name in df.columns:
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)

    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_no_ouliers[column_name] = df_no_ouliers[column_name].apply(lambda x: lower_bound if x < lower_bound
                                                                  else upper_bound if x > upper_bound else x)

  return df_no_ouliers

housing_df_new = remove_ouliers(housing_df)
```

Let's visualise the values of predictors and the target of varaiables in the new dataframe

```python
plot_boxplots(housing_df_new)
```

The diagram above shows the dataset features after replacing the ouliers with the lower and upper bounds to have our data within scope.


Now let's visual the residuals

```python
cols_to_convert = ['MedInc', 'HouseAge', 'AveBedrms', 'Population', 'AveOccup']
housing_df_new[cols_to_convert] = housing_df_new[cols_to_convert].apply(pd.to_numeric, errors='coerce')
housing_df_new['HousePrice'] = pd.to_numeric(housing_df_new['HousePrice'], errors='coerce')

X = sm.add_constant(housing_df_new.drop('HousePrice', axis=1))
y = housing_df_new['HousePrice']

model = sm.OLS(y, X).fit()
```

```python
residuals = model.resid

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
sns.histplot(residuals, kde=True)
plt.title('Residuals Distribution')

plt.subplot(1, 2, 2)
plt.scatter(model.predict(X), residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals vs Predicted')

plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

print("Model coefficient:\n", model.params)
```

The residuals are now more evenly distributed. The coefficients of Average Bedroom shows the most postive value.


Let's now compute the R-squared and the MSE of the linear regressu=ion on the new data

```python
from sklearn.metrics import mean_squared_error, r2_score

predictions = model.predict(X)

print("R-squared: ", r2_score(y, predictions))
print("MSE:", mean_squared_error(y, predictions))
```

We can see that without ouliters the R-quared increase to 0.618 indicating a better-fit while MSE decrease to 0.49 showing an improved accuracy.


**Ouliers removal notably improve the model accuracy**
